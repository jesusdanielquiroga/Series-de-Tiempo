```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ejemplo serie de tiempo con Pandas
# Analizamos una serie de tiempo de pasajeros de una aerolínea
ts = pd.read_csv("https://raw.githubusercontent.com/jesusdanielquiroga/Series-de-Tiempo/main/airline-passengers.csv",
                 squeeze=True,
                 index_col=0,
                 parse_dates=True)
```

```python
# filtrando sólo del 1949-01-01 al 1949-05-01
ts['1949-01-01':'1949-05-01']
```

```python
# valores al 1949-01
ts['1949-01']
```

```python

# graficando
plt.plot(ts['1949'].values, label="1949")
plt.plot(ts['1950'].values, label="1950")
plt.plot(ts['1951'].values, label="1951")
plt.legend()
plt.show()
```

```python

# graficando
df = pd.DataFrame()
df['1949'] = ts["1949"].values
df['1950'] = ts["1950"].values
df['1951'] = ts["1951"].values

df.plot()
```

```python
## graficando en boxplot
df.boxplot()
```
